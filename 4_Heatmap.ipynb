```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_fifa = pd.read_csv('fifa.csv',index_col=0)
df_fifa = df_fifa['2015-01-01':'2017-01-01']
df_fifa
```

```python
plt.figure(figsize=(12,7))
sns.lineplot(data=df_fifa)
plt.xlabel('Kun')
plt.xticks(rotation=55)
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.lineplot(data=df_fifa[['ARG','BRA']])
plt.grid(linestyle='--')
plt.xlabel('Kun')
plt.xticks(rotation=55)
plt.show()
```

```python
df_medals = pd.read_excel('Medals.xlsx')
df_medals = df_medals[:20]
df_medals
```

```python
sns.barplot(x=df_medals['Team/NOC'],y=df_medals['Gold'])
plt.grid(linestyle='--')
plt.xlabel('Teams')
plt.ylabel('Medallar')
plt.xticks(rotation=90)
plt.savefig('medal.png')
plt.show()
```

```python
df_data = pd.read_csv("fifa.csv",index_col=0,parse_dates=[0])
df_data = df_data["2015-01-01":"2018-01-01"]
df_data
```

```python
df_data["oylar"] = [i.month for i in df_data.index]
df_data["yil"]   = [i.year for i in df_data.index]
df_data
```

```python
df_data.index
```

```python
df_data2 = df_data.groupby(['oylar','yil']).max()
df_data2
```

```python
df_data3 = df_data2.unstack(level=1)
df_data3
```

```python
plt.figure(figsize=(12,7))
sns.heatmap(df_data3)
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.heatmap(df_data3,cmap='Blues_r',square=True,linewidths=1,annot=True)
plt.show()
```

```python
data_parvoz = pd.read_csv('flight_delays.csv')
data_parvoz
```

```python
sns.barplot(x=data_parvoz['Month'],y=data_parvoz['US'])
plt.show()
```

```python
cat_data = pd.melt(data_parvoz,id_vars=['Month'],value_vars=['AA','B6','US'])
cat_data
```

```python
sns.catplot(data=cat_data,x="Month",y='value',kind='bar',hue='variable')
plt.show()
```

```python
plt.figure(figsize=(5,12))
cat_data2 = cat_data.drop('variable',axis=1)
sns.heatmap(data=cat_data2,cmap='Blues_r',annot=True,square=True,linewidths=1)
plt.show()
```

```python
data_audi = pd.read_csv('audi.csv')
data_audi
```

```python
plt.plot(data_audi['year'],data_audi['price'],'.')
plt.grid(linestyle='--')
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.histplot(data_audi,x='price')
plt.show()
```

```python
sns.displot(data=data_audi,x='price',aspect=2)
plt.grid()
plt.show()
```

```python
sns.displot(data=data_audi,x='price',hue='transmission',aspect=2,multiple='stack')
plt.grid()
plt.show()
```

```python
sns.displot(data=data_audi,x='price',col='transmission',aspect=2,multiple='stack')
plt.grid()
plt.show()
```

```python
sns.displot(data=data_audi,x='price',kde=True,aspect=2)
plt.grid()
plt.show()
```

```python
sns.displot(data=data_audi,x='price',kind='kde',aspect=2)
plt.grid()
plt.show()
```

```python
sns.kdeplot(data=data_audi,x='price',hue='transmission',shade=True)
plt.show()
```
