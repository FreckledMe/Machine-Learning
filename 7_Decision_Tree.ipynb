```python
x_train = data.drop(1,axis=1).values
y_train = data[1]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)
x_train
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_data,x_test,y_data,y_test = train_test_split(x_train,y_train,test_size=0.2,random_state=32)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=4)
```

```python
knn.fit(x_data,y_data)
```

```python
y_bash = knn.predict(x_test)
```

```python
from sklearn.metrics import jaccard_score,confusion_matrix
```

```python
jaccard_score(y_test,y_bash)
```

```python
confusion_matrix(y_test,y_bash)
```

```python
sns.heatmap(confusion_matrix(y_test,y_bash),annot=True)
```
